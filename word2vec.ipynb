```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
import gensim.downloader as api
```

```python
model = api.load("word2vec-google-news-300")
```

```python
model.most_similar('husband')
```

```python
model.similarity('exam','studies')
```

```python
model.doesnt_match(['exam','studies','burger'])
```

Arithmetic operations

```python
vec=model['parents']-model['mother']+model['father']
model.most_similar([vec])
```

```python
vec=model['parents']+model['burger']-model['father']
model.most_similar([vec])
```

```python
vec=model['burger']+model['mother']-model['father']
model.most_similar([vec])
```
